## Creating a simple plot.

First, we create arrays of the values to plot.

We create an array of 500 values from 0 to 2 pi.

```python
x = np.linspace(0, np.pi * 2, 500)
```

We then calculate the y values we wish to plot. In this case, we'll keep it simple and calculate sin of x (in radians).

```python
y = np.sin(x)
```

# Line Plot (The Object-Oriented Interface)

While you can use `plt.plot()`, the modern and highly recommended approach in Matplotlib is the **Object-Oriented (OO) interface**. We create a Figure (`fig`) and an Axes (`ax`) object using `plt.subplots()`.

Prior to calling `plt.show()`, you can call methods on the `ax` object to modify the plot, such as adding a title or changing the axis limits.

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Sin of x") # optional
ax.set_xlim(0, 2 * np.pi) # optional
plt.show()
```

The default behavior of `plot()` is to plot the points and connect them with lines. 

```python
x_val = np.arange(0, 5, 1)
print(x_val)
y_val = np.random.randint(0, 10, 5)
print(y_val)

fig, ax = plt.subplots()
ax.plot(x_val, y_val)
plt.show()
```

# Scatterplot

If you do not want the points to be connected with a line, you can ask for a scatterplot.

```python
fig, ax = plt.subplots()
ax.scatter(x_val, y_val)
plt.show()
```

# Multiple functions or sequences on the same graph

Repeated calls to `ax.plot()` will add lines to the same plot.

```python
x = np.linspace(0, np.pi * 2, 500)
y = np.sin(x)
z = np.tan(x)
w = np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'blue')
ax.plot(x, z, 'red')
plt.show() # the scale of the two functions make this graph undesirable
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'blue')
ax.plot(x, w, 'orange')
plt.show()
```

# changing labels and axis limits

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim(0, 2 * np.pi)
ax.axis('equal')  # aspect ratio # investigate later
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s = 30, marker = 'd', color = 'maroon') # additional options: s is size, marker is shape
ax.scatter(x, w, s = 80, marker = '*', color = 'darkslategrey')
ax.set_xlim([0, .2])
plt.show()
```

# Plot options: figure size, axis, title

```python
np.random.seed(1)
x_rand = np.random.random(100)
y_rand = np.random.random(100)

fig, ax = plt.subplots(figsize = (4,4))  # define the properties of the figure first
ax.scatter(x_rand, y_rand)             # then add content
ax.axis('equal')            # alter properties
ax.set_title('2-dimensional random uniform values')
plt.show()
```

### adding point size and color options (Colormaps) to a scatter plot

```python
z_size = 100 * np.random.randint(1,11,100)  # use for size
w_color = np.random.randint(0,5,100) # use for color group
print(z_size)
print(w_color)
```

```python
# Instead of manually mapping array values to string colors, 
# we can leverage Matplotlib's powerful `cmap` argument.
# We map the property size (s) to a variable z
# We map the color (c) to a numeric variable w and let cmap handle the palette.

fig, ax = plt.subplots(figsize = (8,8))
scatter = ax.scatter(x_rand, y_rand, s=z_size, c=w_color, cmap='viridis', alpha=0.6)
ax.axis('equal')
fig.colorbar(scatter, ax=ax, label='Color Group (w_color)')
plt.show()
```

```python
# you can change the point markers
fig, ax = plt.subplots()
ax.scatter(x_rand, y_rand, s=z_size, c=w_color, cmap='viridis', marker="*", alpha=0.6)
# But you can't map the markers to a variable
plt.show()
```

# Complex layouts with subplots

```python
# Modern Subplots approach using plt.subplots() array unpacking
fig, axes = plt.subplots(2, 3, figsize=(9, 6))

for i, ax in enumerate(axes.flatten()):
    # ax.text adds text, first two will be the xy position, the character string
    ax.text(0.5, 0.5, f'plot {i}', fontsize=14, ha='center')  
    ax.set_xticks([]) # hide ticks for visual clarity
    ax.set_yticks([])

plt.show()
```

# Subplots example

```python
x = np.linspace(0, np.pi * 2, 500)
y = np.sin(x)
z = np.tan(x)
w = np.cos(x)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, y)
axes[0, 0].set_title("sin(x)")

axes[0, 1].plot(x, w)
axes[0, 1].set_title("cos(x)")

axes[1, 0].plot(x, z)
axes[1, 0].set_title("tan(x)")
axes[1, 0].set_ylim(-5, 5) # necessary for tan to look clean

axes[1, 1].plot(x, y, 'red')
axes[1, 1].plot(x, w, 'blue')
axes[1, 1].set_title("sin and cos")

plt.tight_layout() # Good practice to prevent overlapping labels
plt.show()
```

```python
# More Complicated Grids using GridSpec and add_subplot
fig = plt.figure(figsize=(10, 5))
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.4)

ax1 = fig.add_subplot(grid[0, 0])
ax1.scatter(x_rand[0:10], y_rand[0:10])

ax2 = fig.add_subplot(grid[0, 1:3]) # top row, columns 1:3
ax2.hist(x_rand)

ax3 = fig.add_subplot(grid[1, :2]) # bottom row, columns 0:2
ax3.hist(y_rand)

ax4 = fig.add_subplot(grid[1, 2])
ax4.plot(x_rand[0:4], y_rand[0:4], c='r')

plt.show()
```

### Histograms

```python
data = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(data)
plt.show()
```

```python
# alter bins and color options
fig, ax = plt.subplots()
ax.hist(data, bins=25, alpha=0.5, color='steelblue', edgecolor='black')
plt.show()
```

```python
x1 = np.random.normal(0, 0.8, 10000)
x2 = np.random.normal(-2, 1, 10000)
x3 = np.random.normal(3, 2, 10000)

kwargs = dict(histtype='stepfilled', alpha=0.6, density=True, bins=25, edgecolor='black')

fig, ax = plt.subplots()
ax.hist(x1, **kwargs, color='navy')
ax.hist(x2, **kwargs, color="purple")
ax.hist(x3, **kwargs)
plt.show()
```

```python
# 2d histograms
mean = [0, 0]
cov = [[1, 1], 
       [1, 2]]
x_norm, y_norm = np.random.multivariate_normal(mean, cov, 10000).T

fig, ax = plt.subplots()
h = ax.hist2d(x_norm, y_norm, bins=30, cmap='Blues')
cb = fig.colorbar(h[3], ax=ax)
cb.set_label('counts in bin')
plt.show()
```

```python
fig, ax = plt.subplots()
hb = ax.hexbin(x_norm, y_norm, gridsize=30, cmap='Blues')
cb = fig.colorbar(hb, ax=ax, label='count in bin')
plt.show()
```

## two dimensional functions

```python
x_grid = np.linspace(-3, 3, 101)
y_grid = np.linspace(-3, 3, 101)
X, Y = np.meshgrid(x_grid, y_grid)

def g(x, y): 
    return np.exp(-(x**2)) * np.exp(-(y**2))

Z = g(X, Y)
```

```python
fig, ax = plt.subplots()
ax.contour(X, Y, Z)
ax.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.imshow(Z, origin='lower')
fig.colorbar(im, ax=ax)
plt.show()
```

# Plotting Directly from a Pandas Data Frame

```python
ts = pd.Series(np.random.randn(1000),
   index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index = ts.index, columns = list('ABCD'))
df = df.cumsum()
df.plot() # Pandas returns an Axes object automatically
```

```python
df.loc[:,'A':'B'].plot() # plotting only ts A and B
```

```python
df.iloc[0:100, :2].plot() # plotting a subset of the data
```

```python
df2 = pd.DataFrame(np.random.randint(0,10,size = (5, 4)), columns=['CA', 'NY', 'TX', 'OH'], 
                   index = ["Tech","Agriculture","Manufacturing","Service","Other"])
df2.plot.bar()
plt.title("Type by State")
plt.show()
```

### Titanic Dataset

```python
# Using a highly stable remote repository for the dataset to prevent link rot
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
titanic.Survived = titanic.Survived.astype("category")
titanic.Survived = titanic.Survived.cat.rename_categories({0: "died", 1:"survived"})
pd.crosstab(titanic.Sex, titanic.Survived).plot(kind = "bar", stacked = True)
plt.show()
```

```python
titanic["is_child"] = titanic.Age < 8
pd.crosstab(titanic.is_child, titanic.Survived, normalize = 0).plot(kind = "bar", stacked = True)
plt.show()
```

```python
titanic_summary = titanic.groupby(["Sex", "is_child", "Survived"], observed = False)["PassengerId"].count()
row_totals = titanic_summary.unstack().sum(axis = 1)
titanic_summary.unstack().div(row_totals, axis = 0).plot(kind = "bar", stacked = True)
plt.show()
```
